```python
import pandas as pd
```

```python
df_gdp = pd.read_csv('Book2_csv.csv')
```

```python
df_gdp[30:61]
```

```python
df_gdp['GDP (current US$)']
```

```python
df_gdp_new = df_gdp.fillna(0)
```

```python
df_gdp_new
```

```python
gdp_growth_rate = []
for i in range(31,61,10):
    print(i)
    rate = (  df_gdp_new.iloc[i+9]['GDP (current US$)']  -   df_gdp_new.iloc[i]['GDP (current US$)'] )/df_gdp_new.iloc[i]['GDP (current US$)']
    print(rate)
    gdp_growth_rate.append(rate*100)
  
   
    
```

```python
range_year_GR = ["1991-2000" ,"2001-2010 " ,"2011-2020"]
```

```python
range_year_GR
```

```python

gdp_growth_rate
```

```python
dict_growth_rate = {
    
    "Decade":range_year_GR,
    "Gdp_Growth_rate" :gdp_growth_rate
}
```

```python
df_gdp_growth_rate_decade = pd.DataFrame(dict_growth_rate)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_gdp_growth_rate_decade
```

```python
sns.barplot(x="Decade", y="Gdp_Growth_rate", data=df_gdp_growth_rate_decade,
                 palette="Blues_d")
```
